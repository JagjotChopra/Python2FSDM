```python
# Question 1.
class vehicle:
    #constructor of vehicle class
    def __init__(self,name_of_vehicle,max_speed,average_of_vehicle):
        self.vehicle_name=name_of_vehicle
        self.max_speed=max_speed
        self.average_of_vehicle=average_of_vehicle
    # this method is just for display car info
    def display_info(self):
        print(f'Car Name: {self.vehicle_name}, Max Speed: {self.max_speed}, Average of Vehicle : {self.average_of_vehicle} ')

# making an objefct of vehicle class
car1=vehicle('Accord',200,'50')

#calling the display function of vehicle class with the help of car object
car1.display_info()
```

```python
# Question 2.+
class vehicle:
    #constructor of vehicle class
    def __init__(self,name_of_vehicle,max_speed,average_of_vehicle):
        self.vehicle_name=name_of_vehicle
        self.max_speed=max_speed
        self.average_of_vehicle=average_of_vehicle
        
class advanced_vehicle(vehicle):
    def __init__(self,vehicle_name,max_speed,average):
        super().__init__(vehicle_name,max_speed,average)
    def seating_capacity(self,capacity):
        #return f'{self.vehicle_name} {capacity}'
        return self.vehicle_name, capacity
    
car2=advanced_vehicle('Accord',200,'50')

print(car2.seating_capacity(5))
```
